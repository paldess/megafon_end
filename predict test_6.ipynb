```python
# импорт необходимых элементов
import statistics
from scipy.stats import iqr
import json
from datetime import datetime

from save_file_param import open_json
param_json = open_json()

import pandas as pd
import numpy as np
import dask.dataframe as dd
from sklearn.metrics import f1_score
```

```python
# импорт модели

import dill
import warnings
warnings.filterwarnings('ignore')
with open('models/model_pca_dill.pkl', 'rb') as f:
    pipeline = dill.load(f)
pipeline.steps
```

```python
# загрузка датасетов(features.csv и data_test.csv)

data = dd.read_csv('git_ignore_files/features.csv', sep='\t', blocksize=15e6)
data_test = pd.read_csv('git_ignore_files/data_test.csv')
```

```python
# выбор данных из features.csv только id-шников из data_test.csv

list_id_test = list(data_test['id'].unique())
data = data[data['id'].isin(list_id_test)].compute()
```

```python
# объединение датасетов

data = pd.merge_asof(data_test.sort_values('buy_time'), 
                     data.sort_values('buy_time'), on='buy_time', by='id', direction='nearest')
```

```python
# вероятности подключения услуг абонентом

df = pd.DataFrame()
for i in param_json['uslugi']:
    data['vas_id'] = i
    df[i] = pipeline.predict_proba(data)[:, 1]
df.head(10)
```

```python
# рекомендуемая услуга

recomendate = pd.DataFrame()
recomendate['id'] = data['id']
recomendate['рекомендуемая услуга'] = df.idxmax(axis=1)
recomendate
```

```python
# рекомендация услуг в порядке убывания 
#(колонка 1 - наиболее рекомендуемая, колонка 8 - наименее)

from tqdm import tqdm
recomendate_sorted = pd.DataFrame(columns=[1,2,3,4,5,6,7,8])
for i in tqdm(df.index):
    a = pd.DataFrame()
    a['0'] = df.loc[i]
    recomendate_sorted.loc[recomendate_sorted.shape[0]] = a.sort_values('0', ascending=False).index
for i in recomendate_sorted.columns:
    recomendate[i] = recomendate_sorted[i]
recomendate
```
