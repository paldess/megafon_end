```python
# загрузка подготовленных уравненных по классам данных

data = pd.read_csv('git_ignore_files/data_train_merge_120k.csv')
data.head()
```

```python
# преобразование данных для этапа

data_transform = pipeline[:3].fit_transform(data)
data_transform
```

```python
# проверка скора при простой кластеризации

from sklearn.cluster import KMeans
k = KMeans(n_clusters=2).fit(data_transform)

from sklearn.metrics import f1_score
f1_score(data['target'], k.labels_, average='macro')
```

```python
# подключение PCA

from sklearn.decomposition import PCA
data_transform_pca = data_transform.copy()
pca = PCA(n_components=2).fit_transform(data_transform_pca)
pca
```

```python
import matplotlib.pyplot as plt

plt.scatter(pca[:,0], pca[:,1], c=data['target'])
```

```python
pca = PCA(n_components=10).fit(data_transform)
pca.explained_variance_ratio_
```
